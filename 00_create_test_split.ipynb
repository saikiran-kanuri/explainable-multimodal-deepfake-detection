```python
# Run this in a new cell NOW — manual test split
# since preprocessing already ran, we split from processed data instead

from google.colab import drive
drive.mount('/content/drive')

import os, random, shutil
from pathlib import Path

BASE_DIR = "/content/drive/MyDrive/Colab Notebooks/deepfake-project"

# We split from frames/ instead of raw/ since preprocessing is done
splits = ["fake", "fake_with_audio", "real", "real_with_audio"]

for split in splits:
    src_dir  = os.path.join(BASE_DIR, "data/raw", split)
    test_dir = os.path.join(BASE_DIR, "data/test_videos", split)
    os.makedirs(test_dir, exist_ok=True)
    if not os.path.isdir(src_dir): continue
    videos  = list(Path(src_dir).glob("*.mp4"))
    n_test  = max(5, int(len(videos) * 0.15))
    chosen  = random.sample(videos, min(n_test, len(videos)))
    for v in chosen:
        shutil.copy(str(v), os.path.join(test_dir, v.name))  # COPY not move
    print(f"  {split}: {len(chosen)} videos copied to test_videos/")

print("\n✅ Test set ready — copied (not moved) so training data intact")
```
